```python
# ==============================================================================
# 2. Re-indexing Script
# ==============================================================================

def reindex_labels(label_folder: Path, final_map: dict):
    """Iterates through label files and replaces old class IDs with new ones."""
    print(f"\nProcessing labels in: {label_folder}")
    if not label_folder.exists():
        print("Folder not found. Skipping.")
        return

    label_files = list(label_folder.glob("*.txt"))
    if not label_files:
        print("No label files found. Skipping.")
        return

    for label_path in label_files:
        with open(label_path, 'r') as f:
            lines = f.readlines()

        new_lines = []
        mapping_failed = False

        for line in lines:
            parts = line.strip().split()
            if not parts: continue

            try:
                old_class_id = int(parts[0])
                bbox_coords = " ".join(parts[1:])

                new_class_id = final_map.get(old_class_id)

                if new_class_id is not None:
                    new_lines.append(f"{new_class_id} {bbox_coords}")
                else:
                    print(f"ERROR: Missing map for OLD ID {old_class_id} in {label_path.name}. STOPPING.")
                    mapping_failed = True
                    break

            except Exception as e:
                print(f"Error processing line in {label_path.name}: {e}. STOPPING.")
                mapping_failed = True
                break

        if not mapping_failed:
            with open(label_path, 'w') as f: # Overwrite the file
                f.write("\n".join(new_lines) + "\n")


# --- Execution ---
if __name__ == "__main__":

    # ‚ö†Ô∏è WARNING: BACK UP YOUR LABEL FILES BEFORE EXECUTING THIS!

    for folder in LABEL_FOLDERS_TO_PROCESS:
        reindex_labels(folder, FINAL_MAP)

    print("\nLabel re-indexing complete (or skipped if errors occurred).")
```
